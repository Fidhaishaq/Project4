```python
data=pd.DataFrame(iris.data,columns=iris.feature_names) # we are making it as a dataframe by using feature names as coloumns.
data
```

```python
data['target']=iris.target # we are making a  new coloumn for target values.
data
```

```python
# making an empty list named tvalues and assigning values into it by for loop.
# assigning i as data in the target and searching for the condition.

tvalues=[]
for i in data['target']:
    if i==0:
        tvalues.append('setosa')
    elif i==1:
        tvalues.append('versicolor')
    else:
        tvalues.append('virginica')
```

```python
tvalues
```

```python
data['species']=tvalues # making a new coloumn of spwcies justb for understanding.
data
```

```python
x=iris.data
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2) 
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape 
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(xtrain,ytrain) 
model.score(xtest,ytest)
```

```python
model.predict(xtest)
```

```python
pred=model.predict([[5.2,4.5,3.2,4.9]])
if pred==0: #setosa', 'versicolor', 'virginica'
    print('The species is setosa')
elif pred==1:
    print('The species is versicolor')
else:
    print('The species is virginica')

```
